```python
def remove_outliers(column_name, data_frame):
    '''removes outliers based on the 1.5IQR rule
    
    This ends up removing too many data points, so removed manually
    '''
    
    Q3 = data_frame[column_name].quantile(.75)
    Q1 = data_frame[column_name].quantile(.25)
    IQR = (Q3 - Q1)
    data_frame = data_frame[(data_frame[column_name] < (Q3 + 1.5*IQR)) & 
                                                  (data_frame[column_name] > (Q1 - 1.5*IQR))]
    print(column_name, Q1-1.5*IQR, Q3+1.5*IQR, IQR)
    return data_frame
```

```python
# code attributed to https://deparkes.co.uk/2016/11/04/sort-pandas-boxplot/
def boxplot_sorted(df, by, column, rot=0):
    # use dict comprehension to create new dataframe from the iterable groupby object
    # each group name becomes a column in the new dataframe
    df2 = pd.DataFrame({col:vals[column] for col, vals in df.groupby(by)})
    # find and sort the median values in this new dataframe
    meds = df2.median().sort_values(ascending=False)
    # use the columns in the dataframe, ordered sorted by median value
    # return axes so changes can be made outside the function
    return df2[meds.index].boxplot(rot=rot, return_type="axes")
     
axes = boxplot_sorted(train, by=["Neighborhood"], column="SalePrice")
axes.set_title("Boxplot of Neighborhoods")
plt.xticks(rotation=70)
```

# Fill Missing Values

Normally, it would be better to create new features first before filling in NaN values especially if imputing. This would help avoid any bias in the new features from any imputation steps. But in this case, filling in NaN values is equivalent to data cleaning since missing values are known.

If XGBoost is used, this step might not be necessary.

```python
train.isna().sum()
```

```python
# This cell replaces null values in the training and test set

def null_values(data_frame):
    
    """
    replace null values with the correct values based on missing values in training set
    
    args:
        data_frame - pandas data frame
        
    returns:
        data_frame - pandas data frame with null values filled
    
    """
    # create garage exists feature to set GarageYrBlt for missing values to 
    # this is an alternative to imputing the GarageYrBlt feature
    data_frame['GarageNotExists'] = 0
    data_frame.loc[data_frame['GarageYrBlt'].isna(), 'GarageNotExists'] = 1
    data_frame['GarageYrBlt'].fillna(0, inplace=True)

    data_frame['MasVnrNotExists'] = 0
    data_frame.loc[data_frame['MasVnrArea'].isna(), 'MasVnrNotExists'] = 1
    data_frame['MasVnrArea'].fillna(0, inplace=True)

    # inpute the missing categorical variable values
    # if using minmax scaling (rather than centering data), can impute missing values first
    data_frame['LotFrontage'].fillna(0, inplace=True)
    data_frame['Alley'].fillna('NoAccess', inplace=True)
    data_frame['BsmtQual'].fillna('NoBasement', inplace=True)
    data_frame['BsmtCond'].fillna('NoBasement', inplace=True)
    data_frame['BsmtExposure'].fillna('NoBasement', inplace=True)
    data_frame['BsmtFinType1'].fillna('NoBasement', inplace=True)
    data_frame['BsmtFinType2'].fillna('NoBasement', inplace=True)
    data_frame['FireplaceQu'].fillna('NoFireplace', inplace=True)
    data_frame['GarageType'].fillna('NoGarage', inplace=True)
    data_frame['GarageFinish'].fillna('NoGarage', inplace=True)
    data_frame['GarageQual'].fillna('NoGarage', inplace=True)
    data_frame['GarageCond'].fillna('NoGarage', inplace=True)
    data_frame['PoolQC'].fillna('NoPool', inplace=True)   
    data_frame['Fence'].fillna('NoFence', inplace=True)
    data_frame['MiscFeature'].fillna('MiscFeature', inplace=True)
    data_frame['MasVnrType'].fillna('NoMasonry', inplace=True)
    data_frame['Electrical'].fillna('Missing', inplace=True)
    
    return data_frame

train = null_values(train)
test = null_values(test)

def fill_test_null(data_frame):
    
    """
    fills in missing data in test data set that were missed by the null_values method
    
    args:
        data_frame - pandas data frame (test set)
        
    returns:
        data_frame - pandas data frame without missing values in test set
    """
    
    # TODO: can improve this with more exploration or KNN to find similar rows
    # TODO: could add this back to the test set first and then do imputation
    # TODO: could use XGBoost so not to have to do all of this and missing values taken care of 
    data_frame['MSZoning'].fillna('', inplace=True)
    data_frame['Utilities'].fillna('', inplace=True)
    data_frame['Exterior1st'].fillna('Other', inplace=True)
    data_frame['Exterior2nd'].fillna('Other', inplace=True)
    data_frame['BsmtFinType1'].fillna('NoBasement', inplace=True)
    data_frame['BsmtFinType2'].fillna('NoBasement', inplace=True)
    data_frame['BsmtUnfSF'].fillna(0, inplace=True) # this house has no basement
    data_frame['TotalBsmtSF'].fillna(0, inplace=True) # this house has no basement (could double check other Basement variables)
    data_frame['BsmtFullBath'].fillna(0, inplace=True)
    data_frame['BsmtHalfBath'].fillna(0, inplace=True)
    data_frame['GarageCars'].fillna(0, inplace=True)
    data_frame['GarageArea'].fillna(0, inplace=True)
    data_frame['BsmtFinSF1'].fillna(0, inplace=True)
    data_frame['BsmtFinSF2'].fillna(0, inplace=True)

    data_frame['SaleType'].fillna('Con', inplace=True) # TODO: more exploration to figure this missing value out
    data_frame['KitchenQual'].fillna('TA', inplace=True) # TODO: more exploration to figure this missing value for this house
    data_frame['Functional'].fillna('Mod', inplace=True) # TODO: more exploration to figure this missing value for these houses

    data_frame.loc[data_frame['MSZoning'] == '', 'MSZoning'] = 'RL' # TODO: need to check that this is an appropriate value based on the house type
    data_frame.loc[data_frame['Utilities'] == '', 'Utilities'] = 'AllPub' # mode from training
    data_frame.loc[data_frame['Exterior1st'] == 'Other', 'Exterior1st'] = 'VinylSd' # mode from training

    return data_frame

test = fill_test_null(test)
```

# One Hot Encoding of Categorical Variables

```python
cat_columns = []
for column in train.columns:
    if train[column].dtype == 'object':
        cat_columns.append(column)
```

```python
dummies = pd.get_dummies(train.drop('SalePrice', axis=1).append(test), drop_first=True, columns=cat_columns)
```

```python
train_dummies = dummies.iloc[0:train.shape[0],:].copy()
test_dummies = dummies.iloc[train.shape[0]:,:].copy()
```

```python
train_dummies.loc[:,'SalePrice'] = train.loc[:,'SalePrice']
```

# Create New Features

Besides the GarageExists features, other features can be made. If there had been missing continuous values, it would be better to create new features and then impute. Otherwise, you would have problems with the features being biased from the imputations. Interactions between categorical variables can also be used although this should happen before one hot encoding

```python
# TODO: Turn this into a function to make same features for both sets
# TODO: Create more features/feature_interactions
# TODO: mean encoding especially by neighborhood, but do this with cross validation to avoid overfitting

train_dummies['LotFrontage_LotArea'] = train_dummies['LotFrontage'] / train_dummies['LotArea']
test_dummies['LotFrontage_LotArea'] = test_dummies['LotFrontage'] / test_dummies['LotArea']


train_features = train_dummies.drop(['SalePrice', 'Id'], axis=1)
train_response = train_dummies['SalePrice']
test_features = test_dummies.drop(['Id'], axis=1)
```

```python
# convert data types to integer
for column in train_features.columns:
    train_features[column] = train_features[column].astype(float)
    
for column in test_features.columns:
    test_features[column] = test_features[column].astype(float)
```

# Lasso Regression

```python
random_state = 5
parameters = {'lasso__alpha':[1e-1, 1, 1e1, 1e2, 1e3, 1e4]}
scaler = preprocessing.MinMaxScaler()
lasso = linear_model.Lasso(max_iter=10000)
ml_pipeline = Pipeline([('scale', scaler), ('lasso', lasso)])
gridsearch = GridSearchCV(ml_pipeline, parameters, cv=5, scoring='neg_mean_squared_error', return_train_score=True)
gridsearch.fit(train_features, train_response)
```

```python
# Best root mean squared error score
print('cross-validation rmse', (-gridsearch.best_score_) ** .5)

# Train score
print('training rmse', mean_squared_error(gridsearch.predict(train_features), train_response) ** .5)
```

```python
lasso_model = gridsearch

# best estimator from gridsearch
gridsearch.best_estimator_.named_steps['lasso']
```

```python
# Lasso regression coefficients
list(zip(train_features.columns, lasso_model.best_estimator_.named_steps['lasso'].coef_))

nonzero_features = []
for feature in list(zip(train_features.columns, lasso_model.best_estimator_.named_steps['lasso'].coef_)):
    if feature[1] > 0:
        nonzero_features.append(feature[0])
```

```python
# store predictions
results = pd.DataFrame({'ID':test.Id, 'SalePrice':gridsearch.predict(test_features)})
results.to_csv('results.csv', index=False)
```

# XGBoost

```python
random_state = 5
parameters = {'xgboost__learning_rate':[.2],
             'xgboost__max_depth':[4], 'xgboost__subsample':[.9,1], 'xgboost__colsample_bytree':[.3,.6,.9,1]}
scaler = preprocessing.MinMaxScaler()

xgbregressor = XGBRegressor(random_state=random_state, n_estimators=30)

xgbr_pipeline = Pipeline([('scale', scaler), ('xgboost', xgbregressor)])
gridsearch = GridSearchCV(xgbr_pipeline, parameters, cv=10, scoring='neg_mean_squared_error', return_train_score=True)
gridsearch.fit(train_features.loc[:,nonzero_features], train_response)
```

```python
# Best root mean squared error score
print('cross-validation rmse', (-gridsearch.best_score_) ** .5)

# Train score
print('training rmse', mean_squared_error(gridsearch.predict(train_features.loc[:,nonzero_features]), train_response) ** .5)
```

```python
xgb_model = gridsearch

# best estimator from gridsearch
gridsearch.best_estimator_.named_steps['xgboost']
```

```python
# store predictions in a csv file
results = pd.DataFrame({'ID':test.Id, 'SalePrice':gridsearch.predict(test_features.loc[:,nonzero_features])})
results.to_csv('results.csv', index=False)
```

```python
# SVM
random_state = 5
parameters = {'svm__C':[1e7],
             'svm__gamma':[1e-2]}

scaler = preprocessing.MinMaxScaler()
svm = SVR()

svm_pipeline = Pipeline([('scale', scaler), ('svm', svm)])
gridsearch = GridSearchCV(svm_pipeline, parameters, cv=5, scoring='neg_mean_squared_error', return_train_score=True)
gridsearch.fit(train_features.loc[:,nonzero_features], train_response)
```

```python
# Best root mean squared error score
print('cross-validation rmse', (-gridsearch.best_score_) ** .5)

# Train score
print('training rmse', mean_squared_error(gridsearch.predict(train_features.loc[:,nonzero_features]), train_response) ** .5)
```

```python
svm_model = gridsearch

# best estimator from gridsearch
gridsearch.best_estimator_.named_steps['svm']
```

# KNN

```python
# Best root mean squared error score
print('cross-validation rmse', (-gridsearch.best_score_) ** .5)

# Train score
print('training rmse', mean_squared_error(gridsearch.predict(train_features), train_response) ** .5)
```

```python
knn_model = gridsearch

# best estimator from gridsearch
gridsearch.best_estimator_.named_steps['knn']
```

# Average Results of Different Models

```python
comparison = pd.DataFrame({'lasso':lasso_model.predict(test_features), 
             'xgb':xgb_model.predict(test_features.loc[:,nonzero_features]),
                          'svm':svm_model.predict(test_features.loc[:,nonzero_features])})
comparison['mean'] = comparison.lasso/3 + comparison.xgb/3 + comparison.svm/3
comparison.plot.scatter(x='lasso', y='xgb')
comparison.plot.scatter(x='lasso', y='svm')
comparison.plot.scatter(x='lasso', y='mean')
comparison.plot.scatter(x='svm', y='mean')
comparison.plot.scatter(x='xgb', y='mean')
```
